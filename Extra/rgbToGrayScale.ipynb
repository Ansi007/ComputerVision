```python
import numpy as np
import matplotlib.pylab as plt
im = plt.imread('cat.jpg')
fig = plt.figure(figsize=(10,10))
fig.add_subplot(1,2,1)
plt.title("Before")
plt.imshow(im)

x,y = im.shape[0],im.shape[1]
grayIm = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        grayIm[i,j] = im[i,j].mean()
fig.add_subplot(1,2,2)
plt.title("After")
plt.imshow(grayIm,cmap = 'gray')
```
