# Histogram

```python
import matplotlib.pylab as plt
import numpy as np
im = plt.imread("cat.jpg")
plt.imshow(im)
```

```python
fig = plt.figure(figsize=(10,10))
fig.add_subplot(1,3,1)
plt.title("Objet")
plt.imshow(im)

fig.add_subplot(1,3,2)
plt.title("Rotated by 90")
plt.imshow(np.rot90(im))

fig.add_subplot(1,3,3)
plt.title("Rotated By 180")
plt.imshow(np.rot90(np.rot90(im)))

# fig.add_subplot(2,3,1)
# x = int(input("how many times rotate? ")
# plt.title("Rotated By ",x)
# for i in range(x):
#     im = np.rot90(im)
# plt.imshow(im)
    
```

```python
im_R,im_G,im_B = im.copy(),im.copy(),im.copy()
im_R[:,:,(1,2)] = 0
im_G[:,:,(0,2)] = 0
im_B[:,:,(0,1)] = 0

im_rgb = np.concatenate((im_R,im_G,im_B),axis=1)
plt.figure(figsize=(15,15))
plt.imshow(im_rgb)
```

```python
im = plt.imread("cat.jpg")
plt.hist(np.ravel(im),bins = 255)
plt.show()
```

```python
imR,imG,imB = im[:,:,0],im[:,:,1],im[:,:,2]
fig = plt.figure(figsize = (10,3))
fig.add_subplot(1,3,1)
plt.title("Red")
plt.hist(np.ravel(imR),bins = 255)
plt.show()

fig.add_subplot(1,3,2)
plt.title("Green")
plt.hist(np.ravel(imG),bins = 255)
plt.show()
```
