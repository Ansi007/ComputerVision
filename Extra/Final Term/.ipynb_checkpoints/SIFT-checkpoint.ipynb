# SIFT (Scale Invariant Feature Transform)

+ **Feature** is a piece of information about the content of an image; typically about whether a certain region of the image has certain properties. Features may be specific structures in the image such as points, edges or objects.
+ Finding these image features is called **Feature Detection**.
+ Description of the region around the feature so that it can be found in other images, is called **Feature Description**.
+ Once you have the features and its description, you can 
    + Find same features in all images (correspondence problem)
    + Align them
    + Stitch them together (Jigsaw Puzzle) or do whatever you want.
    


+ A corner is still a corner in a rotated image.
+ A corner may not be a corner if the image is scaled. Harris corner detection is not scale invariant.
+  Scale Invariant Feature Transform (SIFT) extracts keypoints and compute its descriptors.

### SIFT in OpenCV

```python
# Importing required libraries

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
import os

os.makedirs("./Results", exist_ok=True)
```

```python
cv.__version__
```

```python
# Read image
img = cv.imread('book.png')

# Convert to greyscale
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
# Display Image
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.show()
```

#### sift.detectAndCompute() function 
- Finds the keypoint in the images. 
- You can pass a mask if you want to search only a part of image. 
- Each keypoint is a special structure which has many attributes like its 
    - (x,y) coordinates, size of the meaningful neighbourhood, 
    - angle which specifies its orientation, 
    - response that specifies strength of keypoints etc.
    


```python
sift = cv.SIFT_create()
key_points, descriptors = sift.detectAndCompute(gray, None)
```

```python
# Here key_points will be a list of keypoints and descriptors is a numpy array of shape (Number of Keypoints)×128.
# keypoint descriptor is created as: 16x16 neighbourhood around the keypoint is taken. 
# It is divided into quadrants of 4x4 size. For each quadrant, 8 bin orientation histogram is created. 
# So a total of 128 bin values are available.

print(len(key_points))
print(descriptors.shape)
```

#### cv.drawKeyPoints() function  
draws the small circles on the locations of keypoints.

<br/>
<br/>

**drawKeypoints(input_image, key_points, output_image, colour, flag)**

parameters:

- input_image : The image  which  is turned into grayscale and then the key points are extracted using the  SURF  or SIFT algorithms is called input image.
- key_points : The key points obtained from the input picture after using the algorithms are referred to as keypoints.
- output_image :  image on which the keypoints are drawn.
- colour : the colour of the keypoints.
- flag : drawing features are represented by the flag.

```python
dest = cv.drawKeypoints(gray, key_points, gray)
cv.imwrite('./Results/sift_result.jpg', dest)
```

```python
# Display Image
plt.figure(figsize=(8,8))
plt.imshow(dest)
plt.title('SIFT Keypoints')
plt.xticks([])
plt.yticks([])
plt.show()
```

##### Rich Keypoints

If you pass a flag, cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS to it, it will draw a circle with size of keypoint and it will even show its orientation.

```python
img2 = cv.drawKeypoints(gray, key_points, gray, flags = cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv.imwrite('./Results/sift_RichKeypoints.jpg', img2)
```

```python
# Display Image
plt.figure(figsize=(8,8))
plt.imshow(img2)
plt.title('SIFT Rich Keypoints')
plt.xticks([])
plt.yticks([])
plt.show()
```
