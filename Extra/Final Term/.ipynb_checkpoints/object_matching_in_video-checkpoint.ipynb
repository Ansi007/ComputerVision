#### Importing required libraries 

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# read image
img1 = cv.imread('dog.png')  
```

```python
# Display Image
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Convert image to gray scale

img1_gray = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
```

```python
# Display Image
plt.imshow(img1_gray, cmap='gray')
plt.title('Gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# feature matching
# BFMatcher (brute force match)
bf = cv.BFMatcher(cv.NORM_L1, crossCheck = True)
```

**ORB (Oriented FAST and Rotated Brief)**

ORB is a very effective way of detecting the features of the image when compared to SIFT and SURF. ORB is programmed to find fewer features in the image when compared to the SIFT and SURF algorithm because it detects the very important features in less time than them yet this algorithm is considered as a very effective algorithm when compared to other detecting algorithms.
    


```python
#sift
sift = cv.ORB_create(nfeatures=70)

kp_1, des_1 = sift.detectAndCompute(img1_gray, None)
```

```python
video_watch = cv.VideoCapture("video.mp4")


while video_watch.isOpened():
    ret, frame = video_watch.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame. Exiting!!!")
        break
        
        
    try:
        # Convert to graycsale
        img2_gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

        kp_2, des_2 = sift.detectAndCompute(img2_gray, None)

        matches = bf.match(des_1, des_2)
        matches = sorted(matches, key = lambda x:x.distance)

        # plotted only 55 matches here for claritys sake
        img3 = cv.drawMatches(img1_gray, kp_1, img2_gray, kp_2, matches[:55], None, flags = 2)

        img3 = cv.resize(img3, (1540, 540))
        cv.imshow('Feature Matching', img3)
    
    except:
        print("No match found!")
    
    # setting "q" as exiting 
    if cv.waitKey(1) == ord('q'):
        break
        
video_watch.release()
cv.destroyAllWindows()
```
