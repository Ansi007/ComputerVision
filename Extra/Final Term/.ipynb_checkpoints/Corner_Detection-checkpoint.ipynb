> - More compact summary of an image since corners are fewer than edge pixels.
> - We can use corners as image feature.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Read the image
img = cv2.imread(path+'book.png')
cv2_imshow(img)
```

##### Convert to graycsale

```python
# Convert the image to grayscale 
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img = np.float32(gray_img)
cv2_imshow(gray_img)
```

##### Apply a bilateral filter

This filter is highly effective in noise removal while keeping edges sharp.

```python
smoothed_img = cv2.bilateralFilter(gray_img, 5, 75, 75)
cv2_imshow(smoothed_img)
```

## 1. Harris Corner Detector

```python
corners_img = cv2.cornerHarris(smoothed_img, 3, 3, 0.04)
print("original image siz: ", smoothed_img.shape, "cornerness_img: ", corners_img.shape)
```

```python
cv2_imshow(cv2.cvtColor(corners_img, cv2.COLOR_GRAY2RGB))
```

```python
# Dilate the result to mark the corners
corners_img = cv2.dilate(corners_img, None)
cv2_imshow(corners_img)
```

```python
# Threshold for an optimal value, it may vary depending on the image.
img[corners_img > 0.9*corners_img.max()] = [0, 255, 0] # blue, green, red
cv2.imwrite(path+"/Results/book_result.jpg", img)
cv2_imshow(img)
```

Erosion vs Dilation

```python
coins_img = cv2.imread(path+'coins.png',0)
kernel = np.ones((5,5), np.uint8)
coins_erosion = cv2.erode(coins_img, kernel, iterations=3)
coins_dilation = cv2.dilate(coins_img, kernel, iterations=3)
cv2_imshow(coins_img)
```

```python
cv2_imshow(cv2.hconcat([coins_erosion, coins_dilation]))
```

```python
img_ball = cv2.imread(path+'football.png')
gray_ball = np.float32(cv2.cvtColor(img_ball, cv2.COLOR_BGR2GRAY))
smoothed_ball = cv2.bilateralFilter(gray_ball, 5, 75, 75)
corners_ball = cv2.cornerHarris(smoothed_ball, 5, 5, 0.04)
corners_ball = cv2.dilate(corners_ball, None)
img_ball[corners_ball > 0.003 * corners_ball.max()] = [0, 255, 0]
cv2.imwrite("./Results/football_result.jpg", img_ball)

cv2_imshow(img_ball)
```
