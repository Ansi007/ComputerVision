# Matching SIFT Descriptors in Python

## 1. SIFT Robustness to Rotation

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
import os

os.makedirs("./Results", exist_ok=True)
```

```python
# read images
img1 = cv.imread('book.png')  
img2 = cv.imread('book_rotated.jpg') 
```

```python
plt.subplot(121)
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.title('Rotated')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Convert images to gray scale

img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
```

```python
#sift
sift = cv.SIFT_create()

kp_1, des_1 = sift.detectAndCompute(img1, None)
kp_2, des_2 = sift.detectAndCompute(img2, None)
```

###### BFMatcher (brute force match)

Brute-Force matcher is simple. It takes the descriptor of one feature in first set and is matched with all other features in second set using some distance calculation and the closest one is returned.

For BF matcher, first we have to create the BFMatcher object using cv.BFMatcher(). It takes two optional params. First one is normType. It specifies the distance measurement to be used.

Second param is boolean variable, crossCheck which is false by default. If it is true, Matcher returns only those matches with value (i, j) such that i-th descriptor in set A has j-th descriptor in set B as the best match and vice-versa.

```python
# feature matching
# BFMatcher (brute force match)
bf = cv.BFMatcher(cv.NORM_L1, crossCheck = True)

matches = bf.match(des_1, des_2)
matches = sorted(matches, key = lambda x:x.distance)

# plotted only 55 matches here for clarity’s sake
img3 = cv.drawMatches(img1, kp_1, img2, kp_2, matches[:55], None) #cv::DrawMatchesFlags::NOT_DRAW_SINGLE_POINTS = 2
```

```python
# Display Image
plt.figure(figsize=(10, 10))
plt.imshow(img3)
plt.title('SIFT descriptor matching shows robustness to rotation')
plt.xticks([])
plt.yticks([])
plt.savefig("./Results/sift_matches_book_and_rotated.png")
plt.show()
```
