```python
from tensorflow import keras


model = keras.models.load_model(model_path+'TL_model.h5')
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
name_mapping_dict = {'Maria Sharapova': 0,
                     'Kobe Bryant': 1,
                     'Ronaldo': 2,
                     'Kane Williamson': 3}


inv_name_mapping_dict =  {v: k for k, v in name_mapping_dict.items()}

print(inv_name_mapping_dict)
```

```python
img_size = 224
```

**Lets combine it in a function**

```python
def test_func(model, img_size, img_name, class_name_mapping):
  img = plt.imread(img_name)
  img = cv2.resize(img, (img_size, img_size))

  temp_img = np.expand_dims(img, axis=0)
  
  prediction = class_name_mapping[np.argmax(model.predict(temp_img)[0])]
  print("Model prediction: {}".format(prediction))
  
  plt.imshow(img)
  plt.xticks([])
  plt.yticks([])
  plt.show()
```

```python
test_func(model, img_size=224, img_name='1.jpg', class_name_mapping=inv_name_mapping_dict)
```

```python
test_func(model, img_size=224, img_name='2.jpg', class_name_mapping=inv_name_mapping_dict)
```

```python
test_func(model, img_size=224, img_name='3.jpg', class_name_mapping=inv_name_mapping_dict)
```
