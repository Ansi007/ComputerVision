```python
def excel(masterLocation, newLocation):
    if masterLocation == "" or newLocation == "":
        return
    import pandas
    import numpy

    masterFile = pandas.read_excel(masterLocation)
    newFile = pandas.read_excel(newLocation)

    # print whole sheet data
    master = masterFile.to_numpy()
    new = newFile.to_numpy()
    rowsMaster,colsMaster = master.shape
    rowsNew,colsNew = new.shape
    outputArr = []
    for i in range(0,rowsMaster):
        count = 0
        for j in range(0,rowsNew):
            barcodeMaster = master[i][0]
            barcodeNew = new[j][0]
            if barcodeMaster == barcodeNew:
                schoolMaster = master[i][1]
                schoolNew = new[j][1]
                count = count + 1
                if schoolMaster == schoolNew:
                    count = count + 1
                    roomMaster = master[i][2]
                    roomNew = new[j][2]
                    if roomMaster == roomNew:
                        count = count + 1
        if count == 0:
            outputArr.append(numpy.insert(master[i], 0, 'M'))  
        elif count == 1:
            outputArr.append(numpy.insert(master[i], 0, 'C'))
        elif count == 2:
            outputArr.append(numpy.insert(master[i], 0, 'C'))
        elif count == 3:
            outputArr.append(numpy.insert(master[i], 0, 'F'))

    for i in range(0,rowsNew):
        count = 0
        for j in range(0,rowsMaster):
            barcodeNew = new[i][0]
            barcodeMaster = master[j][0]
            if barcodeNew == barcodeMaster:
                count = count + 1
        if count == 0:
            outputArr.append(numpy.insert(new[i], 0, 'N'))

    outputArr = numpy.array(outputArr)
    df = pandas.DataFrame (outputArr, columns =["Status", "Barcode", "School", "Room", "Description", "Manufacture", "Model" , "Cost", "SerialNumber", "Scan Date", "Department", "PO Number", "Replacement Cost"])
    filepath = 'Output.xlsx'
    df.to_excel(filepath, index=False)
    showinfo(
        title = 'Output',
        message = "Output File Generated Successfully"
    )
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog as fd
from tkinter.messagebox import showinfo

# create the root window
root = tk.Tk()
root.title('Excel Scripting')
root.resizable(False, False)
root.geometry('200x100')

masterLocation = ""
newLocation = ""

def select_file_master():
    filetypes = (
        ('Excel files', '*.xlsx'),
        ('All files', '*.*')
    )

    filename = fd.askopenfilename(
        title='Open a file',
        initialdir='/',
        filetypes = filetypes)

    showinfo(
        title='Selected File',
        message = filename
    )
    global masterLocation
    masterLocation = filename
    
def select_file_new():
    filetypes = (
        ('Excel files', '*.xlsx'),
        ('All files', '*.*')
    )

    filename = fd.askopenfilename(
        title='Open a file',
        initialdir='/',
        filetypes = filetypes)

    showinfo(
        title='Selected File',
        message = filename
    )
    global newLocation
    newLocation = filename
    
# open button
open_button_master = ttk.Button(
    root,
    text='Open Master File',
    command = select_file_master
)

open_button_master.pack(expand=True)

# open button
open_button_new = ttk.Button(
    root,
    text='Open New File',
    command = select_file_new
)

open_button_new.pack(expand=True)

def callExcel():
    excel(masterLocation, newLocation)

B = tk.Button(root, text ="Start", command = callExcel)

B.pack()

# run the application
root.mainloop()
```
