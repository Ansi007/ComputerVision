```python
import pyautogui
```

```python
def startKey():
    # Holds down the alt key
    pyautogui.keyDown("alt")
    # Presses the tab key once
    pyautogui.press("shift")
    # Lets go of the alt key
    pyautogui.keyUp("alt")
```

```python
def stopKey():
    # Holds down the alt key
    pyautogui.keyDown("win")
    # Presses the tab key once
    pyautogui.press("r")
    # Lets go of the alt key
    pyautogui.keyUp("win")
```

```python
from tkinter import *
import sys
import time
from PIL import ImageTk, Image

#GLOBAL VARIABLES
global count
count = 0 #0 if 
global popupTimer
popupTimer = 0
global isTimerStarted
isTimerStarted = False
global currentPopUp
currentPopUp = -1

class stopwatch():
    def closeStartPop(self):
        global isTimerStarted
        isTimerStarted = False
        currentPopUp = -1
        self.recPopStart.destroy()
    def closeStopPop(self):
        global isTimerStarted
        isTimerStarted = False
        currentPopUp = -1
        self.recPopStop.destroy()
    def reset(self):
        global count
        count = 1
        self.t.set('00:00:00')
    def start(self):
        global isTimerStarted
        global popupTimer
        global currentPopUp
        if currentPopUp == 1:
            self.closeStopPop()
        global count
        count = 0
        self.timer() 
        startKey()
        self.bt1["state"] = "disabled"
        self.bt2["state"] = "normal"
        #popup_start()
        self.recPopStart = Toplevel(self.root)
        self.recPopStart.geometry("300x100")
        self.recPopStart.title("Child Window")
        self.imgIcon1 = ImageTk.PhotoImage(Image.open("Recording.png"))
        # Create a Label Widget to display the text or Image
        self.imglabel1 = Label(self.recPopStart, image = self.imgIcon1,bg="white")
        self.imglabel1.pack()
        self.imglabel1.place(x=10,y=25)
        Label(self.recPopStart, text= "Recording Started", font=('Mistral 16'),bg="white").place(x=85,y=35)
        windowWidth = self.root.winfo_reqwidth()
        windowHeight = self.root.winfo_reqheight()
        positionRight = int(self.root.winfo_screenwidth() - 310)
        positionDown = int(self.root.winfo_screenheight() - 140)
        self.recPopStart.geometry("+{}+{}".format(positionRight, positionDown))
        self.recPopStart.overrideredirect(True)
        isTimerStarted = True
        popupTimer = 0
        currentPopUp = 0
        self.popupTimer()
        self.recLabel2.place(x=100,y=248)
        self.recPopStart.configure(bg="white")
        self.bt1Off.place(x = 140, y = 248)
    def stop(self):
        global isTimerStarted
        global popupTimer
        global currentPopUp
        if currentPopUp == 0:
            self.closeStartPop()
        global count
        count = 1
        stopKey()
        self.t.set('00:00:00')
        self.bt1["state"] = "normal"
        self.bt2["state"] = "disabled"
        self.recPopStop = Toplevel(self.root)
        self.recPopStop.geometry("300x100")
        self.recPopStop.title("Child Window")
        self.imgIcon2 = ImageTk.PhotoImage(Image.open("Recording.png"))
        # Create a Label Widget to display the text or Image
        self.imglabel2 = Label(self.recPopStop, image = self.imgIcon2,bg="white")
        self.imglabel2.pack()
        self.imglabel2.place(x=10,y=25)
        Label(self.recPopStop, text= "Recording Stopped", font=('Mistral 16'),bg="white").place(x=85,y=35)
        windowWidth = self.root.winfo_reqwidth()
        windowHeight = self.root.winfo_reqheight()
        positionRight = int(self.root.winfo_screenwidth() - 310)
        positionDown = int(self.root.winfo_screenheight() - 140)
        self.recPopStop.geometry("+{}+{}".format(positionRight, positionDown))
        self.recPopStop.overrideredirect(True)
        isTimerStarted = True
        popupTimer = 0
        currentPopUp = 1
        self.popupTimer()
        self.recLabel2.place(x=50000,y=20800)
        self.recPopStop.configure(bg="white")
        self.bt1Off.place(x = 10000, y = 16000)
    def close(self):
        self.root.destroy()
    def popupTimer(self):
        global isTimerStarted
        global currentPopUp
        global popupTimer
        if isTimerStarted == True:
            popupTimer += 1
            self.root.after(1000, self.popupTimer) 
            if popupTimer > 5:
                if currentPopUp == 0:
                    self.closeStartPop()
                else:
                    self.closeStopPop()
                currentPopUp = -1
                isTimerStarted = False
    def timer(self):
        global count
        if(count == 0):
            self.d = str(self.t.get())
            h,m,s = map(int,self.d.split(":")) 
            h = int(h)
            m = int(m)
            s = int(s)
            if(s < 59):
                s += 1
            elif(s == 59):
                s = 0
                if(m < 59):
                    m += 1
                elif(m == 59):
                    m = 0
                    h += 1
            if(h < 10):
                h = str(0) + str(h)
            else:
                h= str(h)
            if(m < 10):
                m = str(0) + str(m)
            else:
                m = str(m)
            if(s < 10):
                s = str(0) + str(s)
            else:
                s = str(s)
            self.d=h+":"+m+":"+s           
            self.t.set(self.d)
            if(count == 0):
                self.root.after(1000,self.timer)     
    def __init__(self):
        self.root=Tk()
        self.root.title("recordt")
        self.root.geometry("500x450")
        self.t = StringVar()
        self.t.set("00:00:00")
        self.lb = Label(self.root,textvariable=self.t,font=("Times 40 bold"),bg="white")
        self.start_img = PhotoImage(file = "Start.png")
        self.bt1 = Button(self.root,text="Start",command=self.start,font=("Times 12 bold"),image = self.start_img, borderwidth = 0,bg="white")
        self.stop_img = PhotoImage(file = "Stop.png")
        self.bt2 = Button(self.root,text="Stop",command=self.stop,font=("Times 12 bold"),image = self.stop_img, borderwidth = 0,bg="white")
        self.lb.place(x = 190,y = 240)
        self.bt1.place(x = 100, y = 160)
        self.bt2.place(x = 300, y = 160)
        self.bt2["state"] = "disabled"
        self.label = Label(self.root,text="",font=("Times 40 bold"))
        #self.root.configure(bg='white')
        # this removes the maximize button
        self.root.resizable(0,0)
        # Create an object of tkinter ImageTk
        self.img = ImageTk.PhotoImage(Image.open("Logo.png"))
        
        # Create a Label Widget to display the text or Image
        self.imglabel = Label(self.root, image = self.img,bg="white")
        self.imglabel.pack()
        self.imglabel.place(x = 160,y=10)

        # Gets both half the screen width/height and window width/height
        # Gets the requested values of the height and widht.
        windowWidth = self.root.winfo_reqwidth()
        windowHeight = self.root.winfo_reqheight()
        positionRight = int(self.root.winfo_screenwidth()/2.5 - windowWidth/2.5)
        positionDown = int(self.root.winfo_screenheight()/3 - windowHeight/3)
        
        self.recIcon = ImageTk.PhotoImage(Image.open("RecStop.png"))
        # Create a Label Widget to display the text or Image
        self.recLabel= Label(self.root, image = self.recIcon,bg="white")
        self.recLabel.pack()
        self.recLabel.place(x = 140,y=248)
        
        self.recIcon2 = ImageTk.PhotoImage(Image.open("RecStart.png"))
        # Create a Label Widget to display the text or Image
        self.recLabel2= Label(self.root, image = self.recIcon2,bg="white")
        self.recLabel2.pack()
        self.recLabel2.place(x=50000,y=20800)
        # Positions the window in the center of the page.
        self.root.geometry("+{}+{}".format(positionRight, positionDown))
        self.root.configure(bg="white")
        
        self.start_img_Off = PhotoImage(file = "StartOff.png")
        self.bt1Off = Button(self.root,text="Start",font=("Times 12 bold"),image = self.start_img_Off, borderwidth = 0,bg="white")
        self.bt1Off.place(x = 10000, y = 16000)
        self.bt1Off["state"] = "normal"
        
        self.root.mainloop()
a=stopwatch()
```
