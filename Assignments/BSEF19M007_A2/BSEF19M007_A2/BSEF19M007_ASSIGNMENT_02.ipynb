```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math

def noise_reduction(img_gray):
    gaussian_gray = 1/16*(np.array(([1,2,1],[2,4,2],[1,2,1]))) #Gaussian Kernel
    img_smooth = convolution(img_gray,gaussian_gray) #Covolve with Gaussian Kernel
    return img_smooth

def flipHorizontal(kernel):
    i = 0 #first column
    j = kernel.shape[1] - 1 #last column
    while True:
        if i == j: #while center is not reached
            break
        else:
            for c in range(kernel.shape[0]): #for each value in columns
                kernel[c][i], kernel[c][j] = kernel[c][j], kernel[c][i] #Swapping
            i = i + 1 #move to new column from start
            j = j - 1 #come back to a new column from end 
    return kernel

def flipVertical(kernel):
    i = 0 #first row
    j = kernel.shape[0] - 1 #last row
    while True:
        if i == j: #while center is not reached
            break
        else:
            for c in range(kernel.shape[1]): #for each value in rows
                kernel[i][c], kernel[j][c] = kernel[j][c], kernel[i][c] #Swapping
            i = i + 1 #move to the new row from start
            j = j - 1 #come back to a new row from end
    return kernel

def flip(kernel):
    kernel = flipHorizontal(kernel)
    kernel = flipVertical(kernel)
    return kernel

def multiply(A,B): #multiply two matrices, 
    ans = 0
    for i in range(A.shape[0]):
        for j in range(A.shape[0]):
            ans += A[i,j] * B[i,j] #multiply same indexex and add all of them
    return ans

def convolution(image,kernel):
    kernel = flip(kernel)
    tempImg = np.zeros(shape=image.shape) #Convoluted Image
    count = int(kernel.shape[0]/2) #Count of rows & columns to be padded
    for c in range(count): #padding rows and columns
        image = np.concatenate((np.zeros((image.shape[0],1), dtype=int), image), axis=1)
        image = np.concatenate((np.zeros((1,image.shape[1]), dtype=int), image), axis=0)
        image = np.append(image, np.zeros((1,image.shape[1]), dtype=int), axis=0)
        image = np.append(image, np.zeros((image.shape[0],1), dtype=int), axis=1)
    x,y = 0,0 #To store new values
    for i in range(0 + count,image.shape[0] - count):
        for j in range(0 + count, image.shape[1] - count):
            A = image[i-count:i+(count+1),j-count:j+(count+1)] #Extract image sub matrix that is required to multiply
            tempImg[x,y] = multiply(A,kernel) #Multiply it with kernel and store the result in new image
            y = y + 1 #move to the next column
        x = x + 1 #move to the next row
        y = 0 
    return tempImg

def sobelX(img):
    sX = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]]) #Sobel X Filter
    return convolution(img,sX)
        
def sobelY(img):
    sY = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]) #Sobel Y Filter
    return convolution(img,sY)

def gradient_calculation(fx,fy):    
    gradient_magnitude = np.sqrt(np.square(fx) + np.square(fy)) #Computing Magnitude
    gradient_direction = np.arctan2(fy,fx) #Computing Direction
    gradient_direction = np.rad2deg(gradient_direction) #Radian To Degree
    gradient_direction += 180 
    return (gradient_magnitude,gradient_direction)

def non_maximum_suppression(gradient_magnitude,gradient_direction):
    img_maximum = np.zeros(gradient_magnitude.shape)
    r,c = gradient_magnitude.shape
    for i in range(1,r-1): #ignoring boundaries
        for j in range(1,c-1):
            gD = gradient_direction[i,j] #current index value
            before,after = 255,255 
            if (0 <= gD < 22.5 or 337.5 <= gD < 360) or (157.5 <= gD < 202.5): #checking angles
                before,after = gradient_magnitude[i,j-1],gradient_magnitude[i,j+1]
            
            elif (22.5 <= gD < 67.5) or (202.5 <= gD < 247.5):
                before,after = gradient_magnitude[i-1,j+1],gradient_magnitude[i+1,j-1]
            
            elif (67.5 <= gD < 112.5) or (247.5 <= gD < 292.5):
                before,after = gradient_magnitude[i-1,j],gradient_magnitude[i+1,j]
            
            elif (112.5 <= gD < 157.5) or (292.5 <= gD < 337.5):
                before,after = gradient_magnitude[i-1,j-1],gradient_magnitude[i+1,j+1]
            
            if gradient_magnitude[i,j] >= before and gradient_magnitude[i,j] >= after: #if current > both
                img_maximum[i,j] = gradient_magnitude[i,j]
                
    return img_maximum
                
def double_threshold(img,tLow,tHigh,tWeak):
    img_threshold = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j] > tHigh: #if above tHigh make it strong
                img_threshold[i,j] = 255
            elif img[i,j] < tLow:
                img_threshold[i,j] = 0 #if low tHigh make it zero
            else:
                img_threshold[i,j] = tWeak #else make it weak
    return img_threshold


def hysteresis(img,weak):
    img_hy = np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img_hy[i,j] == weak:
                if img_hy[i-1,j-1] == 255 or img_hy[i-1,j] == 255 or img_hy[i+1,j] == 255 or img_hy[i,j-1] == 255 or img_hy[i,j+1] == 255 or img_hy[i+1,j-1] == 255 or img_hy[i+1,j] == 255 or img_hy[i+1,j+1] == 255:
                    img_hy[i,j] = 255  #if any of the neighbour is strong make it strong
    return img_hy

def main():
    img_gray = cv.imread('book.png',0)
    img_smooth = noise_reduction(img_gray)
    fx = sobelX(img_smooth)
    cv.imwrite("BSEF19M007_fX.png",fx)
    fy = sobelY(img_smooth)
    cv.imwrite("BSEF19M007_fY.png",fy)
    img_gradient_magnitude,img_gradient_direction = gradient_calculation(fx,fy)
    cv.imwrite("BSEF19M007_Gradient_Magnitude.png",img_gradient_magnitude)
    img_maxima = non_maximum_suppression(img_gradient_magnitude,img_gradient_direction)
    cv.imwrite("BSEF19M007_NMS.png",img_maxima)
    tWeak = 100
    img_threshold = double_threshold(img_maxima,70,90,tWeak)
    img_hystersis = hysteresis(img_threshold,tWeak)
    cv.imwrite("BSEF19M007_Hystersis.png",img_hystersis)
    plt.imshow(img_hystersis,cmap='gray')
    
if __name__ == "__main__":
    main()
```
