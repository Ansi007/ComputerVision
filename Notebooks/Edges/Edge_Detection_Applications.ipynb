One of the many steps involved during the training of a self-driving vehicle is lane detection.

#### Importing required libraries 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
video_watch = cv2.VideoCapture("road.mp4")


while video_watch.isOpened():
    ret, frame = video_watch.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame. Exiting!!!")
        break
        
        
    # Convert to graycsale
    img_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply Guassian blur to smooth the image
    img_smooth = cv2.GaussianBlur(img_gray, (3, 3), sigmaX = 0, sigmaY = 0) 
    
    # Canny Edge Detection
    edges = cv2.Canny(image = img_smooth, threshold1 = 100, threshold2 = 200)
    #cv2.imshow('frame', edges)
    #cv2.imshow('frame', frame)
    
    im_concat = cv2.hconcat([img_gray, edges])
    im_concat = cv2.resize(im_concat, (1540, 540))
    cv2.imshow('Road Lane Detection', im_concat)
    
    # setting "q" as exiting 
    if cv2.waitKey(1) == ord('q'):
        break
        
video_watch.release()
cv2.destroyAllWindows()
```

# Edge-based Face Recognition

```python
# Read the image
img = cv2.imread('face.jpg')
```

```python
# Plot the image

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Convert to graycsale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Guassian blur to smooth the image
img_smooth = cv2.GaussianBlur(img_gray, (3, 3), sigmaX = 0, sigmaY = 0)
edges1 = cv2.Canny(image = img_smooth, threshold1 = 200, threshold2 = 300)
edges2 = cv2.Canny(image = img_smooth, threshold1 = 100, threshold2 = 200)
edges3 = cv2.Canny(image = img_smooth, threshold1 = 20, threshold2 = 40)

plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])


plt.subplot(222)
plt.imshow(edges1, cmap=plt.get_cmap('gray'))
plt.title('(a)')
plt.xticks([])
plt.yticks([])


plt.subplot(223)
plt.imshow(edges2, cmap=plt.get_cmap('gray'))
plt.title('(b)')
plt.xticks([])
plt.yticks([])

plt.subplot(224)
plt.imshow(edges3, cmap=plt.get_cmap('gray'))
plt.title('(c)')
plt.xticks([])
plt.yticks([])
plt.savefig('./Results/edge-based_face_recognition.png', bbox_inches='tight')
plt.show()
```
