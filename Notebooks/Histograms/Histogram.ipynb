#### Importing required libraries

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
#   to read the images
im = plt.imread("cat.jpg")

#to display the images
plt.imshow(im)
```

#### Rotate Image by 90 degrees

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 3, 1)
plt.title("Original")
plt.imshow(im)

fig.add_subplot(1, 3, 2)
plt.title("Rotated")
plt.imshow(np.rot90(im))

fig.add_subplot(1, 3, 3)
plt.title("Rotated")
plt.imshow(np.rot90(np.rot90(im)))
```

#### Filp Image

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.title("Up down flip")
#   Up down flip
plt.imshow(np.flipud(im))

fig.add_subplot(1, 2, 2)
plt.title("Left Right flip")
#   Left Right Flip
plt.imshow(np.fliplr(im))
```

#### Visualize RGB channels

```python
img_R, img_G, img_B = im.copy(), im.copy(), im.copy()

img_R[:, :, (1, 2)] = 0
img_G[:, :, (0, 2)] = 0
img_B[:, :, (0, 1)] = 0

img_rgb = np.concatenate((img_R,img_G,img_B), axis=0)

plt.figure(figsize=(15, 15))
plt.imshow(img_rgb)
```
